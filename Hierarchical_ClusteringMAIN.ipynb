```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/DATA SCIENCE LAB/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
# y = dataset.iloc[:, 3].values
dataset
```

```python
X
```

```python
# Using the dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

n_clusters: The number of clusters to form. In this case, you have set it to 5, which means the algorithm will aim to find 5 clusters.
affinity: The distance metric used to calculate the linkage between clusters. In this case, 'euclidean' distance is used, which measures the Euclidean distance between two data points.
linkage: The linkage criterion used to determine the distance between clusters. 'ward' linkage minimizes the variance within each cluster and tends to produce more balanced and compact clusters.
After fitting the AgglomerativeClustering model to your data, the fit_predict() method is used to perform clustering and assign cluster labels to each data point. The resulting cluster labels are stored in the y_hc variable.

Keep in mind that the variable X should contain your data points in the appropriate format (e.g., a 2-dimensional array-like object).

Once the clustering is performed and cluster labels are assigned, you can further analyze or visualize the results as needed.

```python
# Visualising the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
